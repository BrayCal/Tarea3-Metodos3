$G(x, x') = \frac{1}{\sqrt{4\pi D^2(t - t')}} \, e^{-\frac{(x - x')^2}{4D^2(t - t')}}$
  
   Según la interpretación siguiente: una fuente de radiación puntual ubicada en el origen y que se emite un pulso de radiación luego de 5 unidades de tiempo. Grafique la función para varios valores de tiempo que usted elija, empezando desde $t=0$

```python
import numpy as np
import matplotlib.pyplot as plt

D = 1.0
t_p = 5.0
x = np.linspace(-10, 10, 500)

def green(x, t, D, t_prime):
    if t <= t_p:
        return np.zeros_like(x)
    return (1 / np.sqrt(4 * np.pi * D**2 * (t - t_p))) * np.exp(-(x**2) / (4 * D**2 * (t - t_prime)))

ts = [0,5.7,6,7,10,12]

plt.figure(figsize=(10, 6))
for t in ts:
    G = green(x, t, D, t_p)
    plt.plot(x, G, label=f't = {t} ')

plt.title('Función de Green para la ecuación de difusión\n(Pulso emitido en t\' = 5)')
plt.xlabel('Posición (x)')
plt.ylabel('G(x, 0)')
plt.legend()
plt.grid(True)
plt.show()
```

    $y^{\prime \prime} +4y =x^2$
  * Se obtiene la función de Green:
  \begin{equation*}
  G(x,\xi )= 2\sum^\infty _{n=1} \frac{sin\,(\pi nx)\, sin\,(\pi n\xi)}{4-n^2\pi ^2}

Volvemos a realizar lo mismo, note que siempre trato de agarrar los números impares para mas placer

```python
import numpy as np
import matplotlib.pyplot as plt

def green_ej2(x, xi, N=20):
  for n in range(1, N+1):
    nu = np.sin(n* np.pi *x)* np.sin(n* np.pi * xi)
    de = 4 - (n**2)* (np.pi**2)
    r = 2*nu/de #queda nude xd
  return r

xi2 = [0.1,0.3,0.5,0.7,0.9]
x_v2 = np.linspace(0,1,1500)
plt.figure(figsize=(10,6))

for xi in (xi2):
  G = green_ej2(x_v2, xi)
  plt.plot(x_v2, G, label=f'ξ = {xi}')

plt.title('Función de Green $G(x, \\xi)$ para $n = 20$ \n (Ejemplo 12.1.7 de Heidy)', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$G(x, \\xi)$', fontsize=12)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Como nos indica el enunciado debemos utilizar scipy.special.spence, lo buscamos en la documentación oficial para poderlo invocar en el codigo: https://docs.scipy.org/doc/scipy-1.16.0/reference/generated/scipy.special.spence.html

La cuál dice lo siguiente:

  "Spence’s function, also known as the dilogarithm. It is defined to be:
  \begin{equation}
  \int_1^z \frac{\log(t)}{1-t}dt
  \end{equation}
for complex , where the contour of integration is taken to avoid the branch cut of the logarithm. Spence’s function is analytic everywhere except the negative real axis where it has a branch cut."

En la nota de abajo nos deja algo importante, dice que hay ua convención la cuál define la función Spence por otra forma de integral, de forma que:

\begin{equation}
-\int_0^z \frac{\log(1-t)}{t}dt
\end{equation}

y esto en el codigo se define como $\text{spence}(1-z)$.

Para graficar seguimos las recomendación dada de usar colormesh

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spence

x = np.linspace(-10, 8, 500)
y = np.linspace(-6, 6, 500)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

dlog = spence(Z)
real = np.real(dlog)
imag = np.imag(dlog)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

pcm1 = ax1.pcolormesh(X, Y, real, shading='auto', cmap='Blues')
fig.colorbar(pcm1, ax=ax1, label='Parte Real')
ax1.set_title('Parte Real del Dilogaritmo', fontsize=14)
ax1.set_xlabel('Re(z)')
ax1.set_ylabel('Im(z)')

pcm2 = ax2.pcolormesh(X, Y, imag, shading='auto', cmap='Blues')
fig.colorbar(pcm2, ax=ax2, label='Parte Imaginaria')
ax2.set_title('Parte Imaginaria del Dilogaritmo', fontsize=14)
ax2.set_xlabel('Re(z)')
ax2.set_ylabel('Im(z)')

plt.tight_layout()
plt.show()
```

La ecuación siguiente es la del los polinomios de Chebyshev (De primer y segundo tipo) siendo estas:
\begin{equation}
T_n=\ _2F_1(-n,n;\frac{1}{2};\frac{1-x}{2}), \quad \quad (18.133)

```python
from scipy.special import chebyt, chebyu, hyp2f1
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 3500)
n = [3,6,9]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 6))

for n in (n):
  T_n = chebyt(n)(x)
  T_n_hp = hyp2f1(-n, n, 0.5, (1 - x)/2)
  U_n = chebyu(n)(x)
  U_n_hp = (n + 1) * hyp2f1(-n, n + 2, 1.5, (1 - x)/2)
  ax1.plot(x, T_n, '.-', label=f'$T_{n}(x)$')
  ax1.plot(x, T_n_hp, 'k--', label='Hipergeométrica')
  ax2.plot(x, U_n, '.-', label=f'$U_{n}(x)$')
  ax2.plot(x, U_n_hp, 'k--', label='Hipergeométrica')

ax1.set_title('Polinomios de Chebyshev Tipo 1', fontsize=14)
ax1.set_xlabel('$x$', fontsize=12)
ax1.set_ylabel('$T_n(x)$', fontsize=12)
ax1.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
ax1.grid(True, linestyle='--', alpha=0.6)

ax2.set_title('Polinomios de Chebyshev Tipo 2', fontsize=14)
ax2.set_xlabel('$x$', fontsize=12)
ax2.set_ylabel('$U_n(x)$', fontsize=12)
ax2.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Invocamos la función de spipy ellipj siguiendo la indicación dada, tome 3 valores para que se note de mejor las identidades mencionadas, por lo que las definimos en for y las ploteamos en graficas. y en debemos dejar un espacio por la fase que no estamos utilizando.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ellipj

m_v = [0.2, 0.5, 0.8]
u = np.linspace(0, 10, 500)

fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

for m in m_v:
    sn, cn, dn, _ = ellipj(u, m)
    ec1 = sn**2 + cn**2
    ax1.plot(u, ec1, label=f'm = {m}')
    ec2 = dn**2 + m * sn**2
    ax2.plot(u, ec2, label=f'm = {m}')

ax1.set_title('Identidad: $\\mathrm{sn}^2(u,k) + \\mathrm{cn}^2(u,k) = 1$', fontsize=14)
ax1.set_xlabel('$u$')
ax1.set_ylabel('Valor')
ax1.axhline(1, color='r', linestyle='--', alpha=1)
ax1.legend()
ax1.grid(True)

ax2.set_title('Identidad: $\\mathrm{dn}^2(u,k) + k^2 \\mathrm{sn}^2(u,k) = 1$', fontsize=14)
ax2.set_xlabel('$u$')
ax2.axhline(1, color='r', linestyle='--', alpha=1)
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Para la ultima identidad es algo mas larga

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ellipj

plt.figure(figsize=(10, 6))
m1 = [0.2,0.5,0.8]
v = 2.0
u = np.linspace(0, 10, 3500)

for m in m1:
  sn_u, cn_u, dn_u, _ = ellipj(u, m)
  sn_v, cn_v, dn_v, _ = ellipj(v, m)
  sn_uv, _, _, _ = ellipj(u + v, m)

  nu = sn_u * cn_v * dn_v + sn_v * cn_u * dn_u
  de = 1 - m * sn_u**2 * sn_v**2
  r = nu / de
  plt.plot(u, sn_uv, '.-', label=f'm={m}')
  plt.plot(u, r, 'k--', label=f'Fórmula larga (m={m})')

plt.title('Identidad para $\\mathrm{sn}(u+v,k)$', fontsize=14)
plt.xlabel('$u$')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
